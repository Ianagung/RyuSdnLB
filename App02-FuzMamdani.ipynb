```python
import csv
import multitimer
# importing pandas module 
import pandas as pd 

#global variable
fn = []
val_cpu = []
val_mem = []
```

```python
#job2 - load csv

def job2():
    ip_port = []
    Server = ["Server1", "Server2", "Server3"]
    filename = "ListFileCpuMem.txt"
    file = open(filename,"r")
    global fn
    global val_cpu
    global val_mem
    val_cpu.clear()
    val_mem.clear()
    fn.clear()
    for x in file:
        fn.append(x.rstrip("\n"))
        #print(x)
    file.close()
    for i in fn:
        # making data frame 
        data = pd.read_csv(i, sep = ';') 

        # calling tail() method  
        # storing in new variable
        # get last line data 
        data_bottom = data.tail(1) 

        # display 
        #data_bottom
        #print(data_bottom)

        value = data_bottom.iloc[0]["cpu"]
        val_cpu.append(float(value))       
        value = data_bottom.iloc[0]["mem"]
        val_mem.append(float(value))
    
    print("cpu1,cpu2,cpu3")
    print(val_cpu)
    print("mem1,mem2,mem3")
    print(val_mem)
```

```python
#job1 - timer

def job1():
    #print("do job1")
    job2()
# This timer will run job() five times, one second apart
#timer = multitimer.MultiTimer(interval=3, function=job1, count=5)
timer = multitimer.MultiTimer(interval=5, function=job1)
# Also, this timer would run indefinitely...
timer.start()
```
