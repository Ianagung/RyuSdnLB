```python
import socket, threading
import pandas as pd 
#Get Http response
#IP = http://pycurl.io/
import pycurl
#from datetime import datetime
import datetime
import time
import multitimer
import pandas as pd
from io import BytesIO
class ClientThread(threading.Thread):
    def __init__(self,clientAddress,clientsocket):
        threading.Thread.__init__(self)
        self.csocket = clientsocket
        self.addr = clientAddress
        print ("New connection added: ", self.addr)
        
    def run(self):
        global resp_time1
        global resp_time2
        global resp_time3
        print ("Connection from : ", self.addr)
        #self.csocket.send(bytes("Hi, This is from Server..",'utf-8'))
        msg = ''
        #statInit = True
        # Creating the first Dataframe using dictionary 
        df1 = pd.DataFrame({"id":[1],"cpu":[20], 
                         "mem":[20]},"rsp_tm":[2]})
        # Print df1 
        print(df1, "\n")
        # saving the dataframe
        filename = str(self.addr[0])+ '.csv'
        df1.to_csv(filename)
        #filename = str(self.addr)+ '2.csv'
        resp_time = 1
        
        while True:
            data = self.csocket.recv(2048)
            msg = data.decode()
            if msg=='bye':
              break            
            print ("from client %s %s" % (self.addr, msg))
            
            x = msg.split(";")
            
            #Get respon time update
            if self.addr[0] = '192.168.8.105':
                resp_time = resp_time1
            elif self.addr[0] = '192.168.8.106':
                resp_time = resp_time2
            elif self.addr[0] = '192.168.8.107':
                resp_time = resp_time3
                
            
            # Creating the Second Dataframe using dictionary 
            df2 = pd.DataFrame({"id":[x[0]],"cpu":[float(x[1])]
                                ,"mem":[float(x[2])],"rsp_tm":[resp_time]})
            df1 = df1.append(df2, ignore_index = True)
            print(df1, "\n")
            # saving the dataframe 
            df1.to_csv(filename ,sep=';')
        print ("Client at ", addr , " disconnected...")

def job():
    c = pycurl.Curl()
    global addr1
    global addr2
    global addr3
    global resp_time1
    global resp_time2
    global resp_time3
    
    buffer = BytesIO()    
    #c.setopt(c.URL, 'http://pycurl.io/')
    now1 = datetime.datetime.now() # time object
    #print("now =", now1)
    print("do curl server1")    
    c.setopt(c.URL, addr1)
    c.setopt(c.WRITEDATA, buffer)
    c.perform()
    c.close()
    now2 = datetime.datetime.now() # time object
    selisih = now2 - now1
    #print("now2 =", now2)
    hasil = selisih.seconds + (selisih.microseconds/1000000)
    print(selisih.seconds,":",selisih.microseconds)
    print('elapsed time server1 = {:07.3f}'.format(hasil))
    resp_time1 = float(hasil)
    
    buffer = BytesIO()    
    #c.setopt(c.URL, 'http://pycurl.io/')
    now1 = datetime.datetime.now() # time object
    #print("now =", now1)
    print("do curl server2")    
    c.setopt(c.URL, addr2)
    c.setopt(c.WRITEDATA, buffer)
    c.perform()
    c.close()
    now2 = datetime.datetime.now() # time object
    selisih = now2 - now1
    #print("now2 =", now2)
    hasil = selisih.seconds + (selisih.microseconds/1000000)
    print(selisih.seconds,":",selisih.microseconds)
    print('elapsed time server2 = {:07.3f}'.format(hasil))
    resp_time2 = float(hasil)
    
    buffer = BytesIO()    
    #c.setopt(c.URL, 'http://pycurl.io/')
    now1 = datetime.datetime.now() # time object
    #print("now =", now1)
    print("do curl server3")    
    c.setopt(c.URL, addr3)
    c.setopt(c.WRITEDATA, buffer)
    c.perform()
    c.close()
    now2 = datetime.datetime.now() # time object
    selisih = now2 - now1
    #print("now2 =", now2)
    hasil = selisih.seconds + (selisih.microseconds/1000000)
    print(selisih.seconds,":",selisih.microseconds)
    print('elapsed time server3 = {:07.3f}'.format(hasil))
    resp_time3 = float(hasil)

# address for curl
addr1 = 'http://192.168.8.105:8000/cgi-bin/serverip.cgi'
addr2 = 'http://192.168.8.106:8000/cgi-bin/serverip.cgi' 
addr3 = 'http://192.168.8.107:8000/cgi-bin/serverip.cgi' 
filename = 'ServerData.csv'
#resp_time global variable dalam seconds
resp_time1 = 1
resp_time2 = 2
resp_time3 = 3

#timer
timer = multitimer.MultiTimer(interval=10, function=job, count=-1)
timer.start()

LOCALHOST = "127.0.0.1"
PORT = 1234
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
server.bind((LOCALHOST, PORT))
print("Server started")
print("Waiting for client request..")
while True:
    server.listen(1)
    clientsock, clientAddress = server.accept()
    newthread = ClientThread(clientAddress, clientsock)
    newthread.start()
```
