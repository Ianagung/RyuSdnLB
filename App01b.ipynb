```python
#Get Http response
#IP = http://pycurl.io/
import pycurl
#from datetime import datetime
import datetime
import time
import multitimer
import pandas as pd
from io import BytesIO



def job():
    c = pycurl.Curl()
    global addr
    global df1
    global filename
    buffer = BytesIO()
    print("I'm working...")    
    #c.setopt(c.URL, 'http://pycurl.io/')
    now1 = datetime.datetime.now() # time object
    print("now =", now1)
    #print("type(now) =", type(now1))
    print("do curl")
    #c.setopt(c.URL, 'http://10.0.2.15')
    
    c.setopt(c.URL, addr)
    c.setopt(c.WRITEDATA, buffer)
    c.perform()
    c.close()
    now2 = datetime.datetime.now() # time object
    selisih = now2 - now1
    print("now2 =", now2)
    #print("type(now) =", type(now2))
    print("selisih :", selisih)
    hasil = selisih.seconds + (selisih.microseconds/1000000)
    print(selisih.seconds,":",selisih.microseconds)
    print(hasil)
    df2 = pd.DataFrame({"id":[1],"resp_time":[float(hasil)]})
    df1=df1.append(df2, ignore_index = True)
    print(df1, "\n")
    # saving the dataframe 
    df1.to_csv(filename ,sep=';')
    #body = buffer.getvalue()
    # Body is a byte string.
    # We have to know the encoding in order to print it to a text file
    # such as standard output.
    #print(body.decode('iso-8859-1'))

df1 = pd.DataFrame({"id":[1,1,1], 
                    "resp_time":[1.5,2,0.5]})
# Print df1 
print(df1, "\n")
# saving the dataframe
addr = 'http://pycurl.io/'    
filename = 'Server1.csv'
df1.to_csv(filename)

# This timer will run job() five times, one second apart
timer = multitimer.MultiTimer(interval=10, function=job, count=10)
# Also, this timer would run indefinitely...
timer.start()

#print("love u")
```
